```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

loading dataset

```python
dataset = pd.read_csv('../dataset/train.csv')
dataset.head(3)
```

```python
dataset.info()
```

```python
plt.figure(figsize=(12, 6))
plt.title('Empty data')
sns.heatmap(dataset.isna(), cmap='cividis')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.title('Male vs Female Age Distribution')
sns.distplot(dataset[dataset['Sex'] == 'male']['Age'].dropna(), label='male', kde=False)
sns.distplot(dataset[dataset['Sex'] == 'female']['Age'].dropna(), label='female', kde=False)
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(20, 6))

ax[0].set_title('Male and Female Distribution')
sns.distplot(dataset[dataset['Sex'] == 'male']['Age'].dropna(),
             label='male survived', 
             kde=False,
             ax=ax[0])
sns.distplot(dataset[dataset['Sex'] == 'female']['Age'].dropna(),
             label='female survived', 
             kde=False,
             ax=ax[0])
ax[0].legend()

ax[1].set_title('Male and Female Survived Distribution')
sns.distplot(dataset[(dataset['Sex'] == 'male') & (dataset['Survived'] == 1)]['Age'].dropna(),
             label='male survived', 
             kde=False,
             ax=ax[1])
sns.distplot(dataset[(dataset['Sex'] == 'female') & (dataset['Survived'] == 1)]['Age'].dropna(),
             label='female survived', 
             kde=False,
             ax=ax[1])
ax[1].legend()


ax[2].set_title('Male and Female not Survived Distribution')
sns.distplot(dataset[(dataset['Sex'] == 'male') & (dataset['Survived'] == 0)]['Age'].dropna(),
             label='male not survived', 
             kde=False,
             ax=ax[2])
sns.distplot(dataset[(dataset['Sex'] == 'female') & (dataset['Survived'] == 0)]['Age'].dropna(),
             label='female not survived', 
             kde=False,
             ax=ax[2])
ax[2].legend()

plt.show()
```
